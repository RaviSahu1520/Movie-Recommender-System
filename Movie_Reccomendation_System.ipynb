# **Movie_Reccomender_System**

```python
import pandas as pd
import numpy as np
import ast

# Load the data from CSV files
credit = pd.read_csv("/content/tmdb_5000_credits.csv")
movie = pd.read_csv("/content/tmdb_5000_movies.csv")

# Merge the two datasets on the 'title' column
movies = movie.merge(credit, on='title')

# Print the shapes of the dataframes to verify the merge
print(credit.shape)
print(movie.shape)
print(movies.shape)

# Select only the relevant columns
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]

# Drop any rows with missing values
movies.dropna(inplace=True)

# Define a function to convert stringified lists into actual lists (for genres, keywords, and cast)
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

# Apply the convert function to the 'genres' and 'keywords' columns
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)

# Define a function to keep only the first 3 cast members
def convert3(obj):
    counter = 0
    L = []
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

# Apply the convert3 function to the 'cast' column
movies['cast'] = movies['cast'].apply(convert3)

# Define a function to extract the director's name
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
      i['job'].lower == 'director'
      L.append(i['name'])
      break
    return L

# Apply the fetch_director function to the 'crew' column and create a new 'director' column
movies['crew'] = movies['crew'].apply(fetch_director)

# Display the first few rows of the final dataframe to check the results
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split() if x is not None else [])
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id' , 'title' , 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# we have to convert all the tags data into vector so we match all the vector .
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# there is a problem in our tags data is that 'loving' , 'loved' , 'love' are consider three different words
# but this thing affects our model so for that we have to convert our data in a proper format
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# now check that our data is clean now and there is no duplicate word in our data
feature_name = cv.get_feature_names_out()
```

```python
# now we have to calculate distance between all the vectors in this case  we calculate cosine distance.
# not euclid distance because in a big data this is not beneficial
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
# def recommend(movie):
#   movie_index = new_df[new_df['title'] == movie].index[0]
#   distances = similarity[movie_index]
#   movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

#   for i in movies_list:
#     print(i[0])
def recommend(movie):
    # Check if the movie exists in the DataFrame
    if movie not in new_df['title'].values:
        print(f"'{movie}' not found in the dataset.")
        return

    # Get the index of the movie in the DataFrame
    movie_index = new_df[new_df['title'] == movie].index[0]

    # Get the similarity scores for the movie
    distances = similarity[movie_index]

    # Enumerate and sort the distances, then get the top 5 similar movies
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    # Print the titles of the recommended movies
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
